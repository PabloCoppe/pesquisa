**Versão Simplificada**

Variáveis de entrada:

f - função objetivo

a - extremo esquerdo do intervalo de inspeção

b - extremo direito do intervalo de inspeção

e - tolerância (critério de parada)

xi é a raiz aproximada obtida pelo método

```python
# importação da biblioteca matemática
import math  
import numpy as np # outra biblioteca matemática
# Definir uma função (escrever neste trecho a função que será utilizada)
def f(x):
      return x**3-9*x+3 
# método da bissecção
def bisec(f,a,b,e):
# Aplicando teorema de Bolzano
  if f(a)*f(b)<0:
      while (math.fabs(b-a)/2 >e):
          xi = (a+b)/2
          if f(xi)==0:
            return "A raiz é: " + str(xi)       
          else:
              if f(a)*f(b)<0:
                b=xi
              else:
                  a=xi
      return "O valor da raiz é: " + str(xi)
  else:
    return "Não há raiz neste intervalo"
```

```python
bisec(f,0,1,0.001)
```

**Versão Amplificada** 

O seguinte código é uma implementação em Python do algoritmo da bisseção. As variáveis de entrada são:

f - função objetivo

a - extremo esquerdo do intervalo de inspeção 

b - extremo direito do intervalo de inspeção 

TOL - tolerância (critério de parada)

N - número máximo de iterações

A variável de saída é:

p - aproximação da raiz de f, isto é, f(x) aproximadamente p

```python
def f2(x):
  return x*np.log(x) -1
def bi(f2,a,b,TOL,N):
  i = 1  
  fa = f2(a)  
  while (i <= N):  
      #iteracao da bissecao  
      p = a + (b-a)/2  
      fp = f2(p)  
      #condicao de parada  
      if ((fp == 0) or ((b-a)/2 < TOL)):  
          return p  
      #bissecta o intervalo  
      i = i+1  
      if (fa * fp > 0):  
          a = p  
          fa = fp  
      else:  
          b = p  

  return NameError('Num. max. de iter. excedido!')
```

```python
bi(f2,0,1,0.1,9)
```
